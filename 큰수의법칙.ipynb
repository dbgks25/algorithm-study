```python
# 데이터 입력
n,m,k = map(int, input().split())
data = list(map(int,input().split()))

# 1, 2번째로 큰 수 가려내기
data.sort()
first = data[n-1]
second = data[n-2]

# 가장 큰 수 개수
firnum = (m//k)*k
# 두 번째로 큰 수 개수
secnum = m//k

# cycle이 아닌 경우에 가장 큰 수 계산 과정
if m%k != 0:
  firnum += m%(k+1)
result = firnum*first + secnum*second
print(result)
```
